**->** **ESCAPE SEQUENCE / CHARACHTERS**

```python
# \n use for new or next line.
print("my name is harshal\nmy age is 22\nmy gender is male")
```

```python
# \' single quote.
print('This is \'code and debug\'')
```

```python
# \" Double quote.
print("Bolo \"Jai shree Ram\"")
```

```python
#\\ Backslash.
print("distance = speed\time")#here \t means tab space which is also an escape sequence.
```

```python
print("Distance = speed\\time") #here we learn use of backslash escape sequence.
```
